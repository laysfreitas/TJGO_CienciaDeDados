## Matrículas Educação Básica De Sexo e Raça/Cor Por Região

```python
# Filtrar os dados para incluir apenas do sexo feminino ---
df_sexo_raca_cor = df_filtrado[['NO_REGIAO',
                            'QT_MAT_BAS_FEM',
                           'QT_MAT_BAS_MASC',
                           'QT_MAT_BAS_ND',
                           'QT_MAT_BAS_BRANCA',
                           'QT_MAT_BAS_PRETA',
                           'QT_MAT_BAS_PARDA',
                           'QT_MAT_BAS_AMARELA',
                           'QT_MAT_BAS_INDIGENA']].copy()

# Agrupar o DataFrame pela coluna 'NO_REGIAO'
df_agrupado = df_sexo_raca_cor.groupby('NO_REGIAO')

# Calcular a média de todas as colunas numéricas para cada região
media_por_regiao = df_agrupado.mean()

# Calcular a mediana de todas as colunas numéricas para cada região
mediana_por_regiao = df_agrupado.median()

# Calcular a moda para cada região
# .iloc[0] pega o primeiro valor da moda, caso haja empates.
moda_por_regiao = df_agrupado.apply(lambda x: x.mode().iloc[0], include_groups=False)


print("--- MÉDIA por Região ---")
display(media_por_regiao.round(2))

print("--- MEDIANA por Região ---")
display(mediana_por_regiao)

print("--- MODA por Região ---")
display(moda_por_regiao)
```
