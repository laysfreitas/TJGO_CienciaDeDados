## Matrículas Educação Básica De Sexo e Raça/Cor Por Região

```python
# Filtrar os dados para incluir apenas do sexo feminino ---
df_sexo_raca_cor = df_filtrado[['NO_REGIAO',
                            'QT_MAT_BAS_FEM',
                           'QT_MAT_BAS_MASC',
                           'QT_MAT_BAS_ND',
                           'QT_MAT_BAS_BRANCA',
                           'QT_MAT_BAS_PRETA',
                           'QT_MAT_BAS_PARDA',
                           'QT_MAT_BAS_AMARELA',
                           'QT_MAT_BAS_INDIGENA']].copy()

# Agrupar o DataFrame pela coluna 'NO_REGIAO'
df_agrupado = df_sexo_raca_cor.groupby('NO_REGIAO')

# Calcular a média de todas as colunas numéricas para cada região
media_por_regiao = df_agrupado.mean()

# Calcular a mediana de todas as colunas numéricas para cada região
mediana_por_regiao = df_agrupado.median()

# Calcular a moda para cada região
# .iloc[0] pega o primeiro valor da moda, caso haja empates.
moda_por_regiao = df_agrupado.apply(lambda x: x.mode().iloc[0], include_groups=False)


print("--- MÉDIA por Região ---")
display(media_por_regiao.round(2))

print("--- MEDIANA por Região ---")
display(mediana_por_regiao)

print("--- MODA por Região ---")
display(moda_por_regiao)
```

```python
# Preparar os dados para o agrupamento
regioes = media_por_regiao.index
x_pos = np.arange(len(regioes)) # Posições numéricas para as regiões [0, 1, 2, 3, 4]
bar_width = 0.35 # Largura das barras

# Criar a figura e os eixos
fig, ax = plt.subplots(figsize=(12, 7))

# Barras para matrículas femininas (deslocadas para a esquerda)
bars_fem = ax.bar(x_pos - bar_width/2, media_por_regiao['QT_MAT_BAS_FEM'], bar_width, label='Feminino', color='lightcoral')
# Barras para matrículas masculinas (deslocadas para a direita)
bars_masc = ax.bar(x_pos + bar_width/2, media_por_regiao['QT_MAT_BAS_MASC'], bar_width, label='Masculino', color='skyblue')

# Adicionar os rótulos
ax.bar_label(bars_fem, padding=3, fmt='%d')
ax.bar_label(bars_masc, padding=3, fmt='%d')

ax.set_xlabel('Região', fontsize=12)
ax.set_ylabel('Média de Matrículas na Educação Básica', fontsize=12)
ax.set_title('Média de Matrículas na Ed. Básica por Região e Gênero', fontsize=16)
ax.set_xticks(x_pos) # Garante que os ticks fiquem no centro dos grupos
ax.set_xticklabels(regioes) # Coloca os nomes das regiões no eixo X
ax.legend() # Mostra a legenda ('Feminino', 'Masculino')

fig.tight_layout() # Ajusta o layout para evitar cortes
plt.show()
```

```python
# Supondo que 'df_filtrado' já exista
# --- CORREÇÃO 1: Selecionar apenas as colunas de etnia para este gráfico ---
colunas_etnia = [
    'NO_REGIAO',
    'QT_MAT_BAS_ND',
    'QT_MAT_BAS_BRANCA',
    'QT_MAT_BAS_PRETA',
    'QT_MAT_BAS_PARDA',
    'QT_MAT_BAS_AMARELA',
    'QT_MAT_BAS_INDIGENA'
]
df_raca_cor = df_filtrado[colunas_etnia].copy()

# Agrupar o DataFrame pela coluna 'NO_REGIAO' e calcular a média
media_por_regiao = df_raca_cor.groupby('NO_REGIAO').mean()

# 1. Preparar os dados: "Derreter" (melt) o DataFrame para o formato longo
df_melted = media_por_regiao.reset_index().melt(
    id_vars='NO_REGIAO',
    var_name='Etnia',
    value_name='Média de Matrículas'
)
# Limpa os nomes para a legenda
df_melted['Etnia'] = df_melted['Etnia'].str.replace('QT_MAT_BAS_', '').str.replace('_', ' ').str.title()


# 2. Criar o gráfico
sns.set_style("whitegrid")
plt.figure(figsize=(15, 8))

# --- CORREÇÃO 2: Usar o nome correto da coluna 'NO_REGIAO' ---
ax = sns.barplot(
    data=df_melted,
    x='NO_REGIAO', # Nome corrigido
    y='Média de Matrículas',
    hue='Etnia' # 'hue' agrupa as barras automaticamente por etnia
)

# 3. Adicionar os rótulos (opcional)
for container in ax.containers:
    ax.bar_label(container, fmt='%d', fontsize=10, padding=3)


# 4. Finalizar o gráfico
plt.xlabel('Região', fontsize=12)
plt.ylabel('Média de Matrículas na Educação Básica', fontsize=12)
plt.title('Média de Matrículas na Ed. Básica por Região e Etnia/Cor', fontsize=16)
plt.legend(title='Etnia/Cor') # Adiciona um título à legenda

plt.tight_layout()
plt.show()
```

```python
# Agrupar pela nova coluna e calcular a média da coluna 'IN_LABORATORIO_INFORMATICA'
# Multiplicamos por 100 para obter o resultado em percentual
media_labs_por_dependencia = df_filtrado.groupby('TP_DEPENDENCIA')['IN_LABORATORIO_INFORMATICA'].mean() * 100

# Ordenar os valores para um gráfico mais claro
media_labs_por_dependencia = media_labs_por_dependencia.sort_values(ascending=False)

print("\nPercentual de escolas com Laboratório de Informática por Dependência:")
print(media_labs_por_dependencia)
```

```python
# Configurar o estilo e o tamanho do gráfico
sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))

# Criar o gráfico de barras
ax = sns.barplot(
    x=media_labs_por_dependencia.index,
    y=media_labs_por_dependencia.values,
    palette='viridis' # Escolha de uma paleta de cores
)

# Adicionar os rótulos de dados (percentuais) em cima de cada barra
for container in ax.containers:
    # O '%.1f%%' formata o número com uma casa decimal e adiciona o símbolo de %
    ax.bar_label(container, fmt='%.1f%%', fontsize=12, padding=3)

# Adicionar títulos e rótulos
plt.title('Percentual de Escolas com Laboratório de Informática por Dependência', fontsize=16)
plt.xlabel('Tipo de Dependência', fontsize=12)
plt.ylabel('Percentual (%)', fontsize=12)
plt.ylim(0, 100) # Define o limite do eixo Y de 0 a 100

# Exibir o gráfico
plt.tight_layout()
plt.show()
```
