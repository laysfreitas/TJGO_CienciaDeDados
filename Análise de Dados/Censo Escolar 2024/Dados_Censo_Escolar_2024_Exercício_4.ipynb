### Análise da Distribuição de Recursos

A infraestrutura de tecnologia (computadores, laboratórios) é distribuída de forma igualitária entre as escolas, ou existem grandes disparidades dependendo da região, localização (urbana/rural) e dependência administrativa?

```python
df_infra_tec = df_filtrado[['TP_DEPENDENCIA','NO_REGIAO','TP_LOCALIZACAO',
                            'QT_DESKTOP_ALUNO','QT_COMP_PORTATIL_ALUNO',
                            'QT_TABLET_ALUNO']]

display(df_infra_tec)
```

```python
agrupado_mean = df_infra_tec.groupby(['NO_REGIAO', 'TP_DEPENDENCIA', 'TP_LOCALIZACAO']).mean().reset_index()
agrupado_mean = agrupado_mean.round(2)
display(agrupado_mean.head())

# Código com rótulos de dados
plt.figure(figsize=(16, 9))
# Usar 'ax' para poder manipular os elementos do gráfico
ax = sns.barplot(data=agrupado_mean, x='NO_REGIAO', y='QT_DESKTOP_ALUNO', hue='TP_DEPENDENCIA')

# Adicionando os rótulos
for container in ax.containers:
    ax.bar_label(container, fmt='%.2f', label_type='edge', size=10, rotation=45, padding=9)

# Melhoras estéticas
ax.set_title('Média de Computadores de Mesa por Região e Tipo de Dependência', fontsize=16)
ax.set_xlabel('Região', fontsize=14)
ax.set_ylabel('Média de Computadores de Mesa', fontsize=14)
ax.tick_params(axis='x', rotation=45, labelsize=11)

# Ajustar o limite do eixo Y para dar espaço aos rótulos
ax.set_ylim(top=ax.get_ylim()[1] * 1.15)

plt.legend(title='Tipo de Dependência')
plt.tight_layout() # Ajusta o layout para evitar que os rótulos cortem
plt.show()
```

O gráfico da Média de Computadores de Mesa por Região e Tipo de Dependência evidencia uma profunda estratificação no acesso à tecnologia. ### Relação entre Quantidade de Alunos e Tecnologia

```python
# Identificação Dados Lineares
# Relação entre Quantidade de Alunos da Educação Básica e Infraestrutura da Tecnologia
df_tec_alunos = df_filtrado[['TP_DEPENDENCIA','NO_REGIAO','TP_LOCALIZACAO',
                            'QT_MAT_BAS','QT_MAT_INF','QT_MAT_FUND','QT_MAT_MED',
                            'QT_DESKTOP_ALUNO','QT_COMP_PORTATIL_ALUNO',
                            'QT_TABLET_ALUNO']]

df_tec_alunos = df_tec_alunos[df_tec_alunos['QT_MAT_BAS'] > 0].copy()

df_tec_alunos['QT_DESKTOP_ALUNO'] = df_tec_alunos['QT_DESKTOP_ALUNO'].fillna(0).astype(int)
df_tec_alunos['QT_COMP_PORTATIL_ALUNO'] = df_tec_alunos['QT_COMP_PORTATIL_ALUNO'].fillna(0).astype(int)
df_tec_alunos['QT_TABLET_ALUNO'] = df_tec_alunos['QT_TABLET_ALUNO'].fillna(0).astype(int)

df_tec_alunos['QT_TEC_TOTAL'] = df_tec_alunos['QT_DESKTOP_ALUNO'] + \
                                  df_tec_alunos['QT_COMP_PORTATIL_ALUNO'] + \
                                  df_tec_alunos['QT_TABLET_ALUNO']

# Exibir as primeiras linhas do DataFrame resultante e suas informações
print("DataFrame df_tec_alunos preparado:")
display(df_tec_alunos)
print("\nInformações do DataFrame df_tec_alunos:")
df_tec_alunos.info()
```

```python
plt.figure(figsize=(12, 7))
sns.scatterplot(
    data=df_tec_alunos[df_tec_alunos['QT_MAT_BAS'] > 0], # Filtra escolas com matrícula para evitar aglomeração em (0,0)
    x='QT_MAT_BAS',
    y='QT_TEC_TOTAL',
    alpha=0.6, # Ajuda a ver a densidade dos pontos
    s=20 # Tamanho dos pontos
)
plt.title('Relação entre Matrículas e Total de Computadores por Escola', fontsize=16)
plt.xlabel('Número de Matrículas na Educação Básica', fontsize=12)
plt.ylabel('Total de Computadores para Alunos', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

O gráfico revela que o cenário educacional brasileiro, no que tange a recursos digitais, é caracterizado por:

Uma grande massa de escolas pequenas com poucos computadores.

Uma relação positiva, porém fraca, entre o número de alunos e a quantidade de computadores, indicando que mais alunos não garante proporcionalmente mais equipamentos.

Uma forte desigualdade de infraestrutura entre escolas de mesmo porte, sendo este talvez o insight mais importante do gráfico.

Em suma, a quantidade de computadores em uma escola é influenciada por muitos outros fatores além do seu tamanho em número de matrículas.

```python
covariancia = df_tec_alunos['QT_MAT_BAS'].cov(df_tec_alunos['QT_TEC_TOTAL'])
print(f"Covariância entre Matrículas e Computadores: {covariancia:.2f}")
```

Este valor, por ser positivo, nos diz a direção da relação: quando uma variável sobe, a outra também tende a subir. Ou seja, mais matrículas está associado a mais computadores.

```python
correlacao_pearson = df_tec_alunos['QT_MAT_BAS'].corr(df_tec_alunos['QT_TEC_TOTAL'])
print(f"Correlação de Pearson entre Matrículas e Computadores: {correlacao_pearson:.2f}")
```

O resultado da covariância (7436.78) mostra que a relação entre o número de matrículas e o total de computadores é positiva. O cálculo da Correlação de Pearson (0.40) confirma essa direção positiva e, mais importante, quantifica a força dessa relação como fraca a moderada.

Isso valida perfeitamente o que vimos no gráfico de dispersão: existe uma tendência de aumento, mas ela não é forte e a grande dispersão dos pontos mostra que o número de matrículas de uma escola não é um fator muito confiável para prever sua quantidade de computadores. A correlação de 0.40 é a expressão numérica exata dessa realidade visual.

Coeficiente de Correlação de Spearman para Dados Não Lineares

O que é o Coeficiente de Spearman?
Ao contrário do Coeficiente de Pearson (que mede a força de uma relação linear), o Coeficiente de Spearman (ρ ou r_s) mede a força e a direção de uma relação monotônica entre duas variáveis.

Relação Monotônica: Significa que, conforme uma variável aumenta, a outra tende a consistentemente aumentar (relação monotônica crescente) ou consistentemente diminuir (relação monotônica decrescente), mas não necessariamente a uma taxa constante (que seria uma relação linear).

Exemplo: Se o número de alunos sempre aumenta quando o número de computadores aumenta, mas não em uma proporção exata.

Como funciona: Spearman não usa os valores brutos das variáveis. Em vez disso, ele classifica os dados (atribui um ranking a cada valor) e, em seguida, calcula a Correlação de Pearson sobre esses ranks. Isso o torna menos sensível a outliers e capaz de capturar relações que não são estritamente lineares.

Valores: Varia de -1 a +1.

+1: Perfeita relação monotônica crescente.

-1: Perfeita relação monotônica decrescente.

0: Ausência de relação monotônica.

```python
correlacao_spearman = df_tec_alunos['QT_MAT_BAS'].corr(df_tec_alunos['QT_TEC_TOTAL'], method='spearman')

print(f"Correlação de Spearman entre Matrículas e Computadores: {correlacao_spearman:.2f}")
```

O resultado da Correlação de Spearman (0.52) nos dá uma visão talvez mais fiel da relação entre as variáveis. Ele sugere que, embora a relação não siga uma linha reta perfeita (por isso a correlação de Pearson é mais baixa), existe uma tendência moderada e consistente de que um aumento no número de alunos esteja associado a um aumento no número de computadores.

```python
# Evitar divisão por zero: substitua 0 computadores por NaN ou um número pequeno para cálculo,
# ou filtre essas escolas para a métrica, mas cuidado para não perder o contexto
df_tec_alunos['ALUNOS_POR_COMPUTADOR'] = df_tec_alunos['QT_MAT_BAS'] / df_tec_alunos['QT_TEC_TOTAL'].replace(0, np.nan) # Substitui 0 por NaN para não dividir por zero

# Se preferir uma análise mais robusta, remova NaNs criados pela divisão por zero para esta métrica
df_com_comp = df_tec_alunos[df_tec_alunos['QT_TEC_TOTAL'] > 0].copy() # Trabalhe com uma cópia para evitar SettingWithCopyWarning
df_com_comp['ALUNOS_POR_COMPUTADOR'] = df_com_comp['QT_MAT_BAS'] / df_com_comp['QT_TEC_TOTAL']

# Analise essa nova métrica por dependência, localização, etc.
media_alunos_por_comp_dependencia = df_com_comp.groupby('TP_DEPENDENCIA')['ALUNOS_POR_COMPUTADOR'].mean().sort_values(ascending=False).round(2)
print("\nMédia de Alunos por Computador, por Dependência:")
print(media_alunos_por_comp_dependencia)
```

### Acesso vs. Infraestrutura - Ter internet significa poder usar?

```python
# Das escolas que possuem acesso à internet, quantas efetivamente a disponibilizam para os alunos? E como isso varia pelo Brasil?
df_internet_aluno = df_filtrado[['NO_REGIAO', 'TP_DEPENDENCIA', 'TP_LOCALIZACAO', 'IN_INTERNET', 'IN_INTERNET_ALUNOS']]
df_internet_aluno_ = df_internet_aluno[df_internet_aluno['IN_INTERNET'] == 1].copy()
display(df_internet_aluno)
```

```python
# Neste novo grupo, calcule a média da coluna IN_INTERNET_ALUNOS (como os valores são 0 ou 1, a média representa o percentual de escolas que oferecem internet aos alunos). Agrupe o resultado por NO_REGIAO.
media_internet_alunos_regiao = df_internet_aluno_.groupby('NO_REGIAO')['IN_INTERNET_ALUNOS'].mean().sort_values(ascending=False).round(4)
print("\nProporção de escolas com internet que disponibilizam para alunos, por Região:")
display(media_internet_alunos_regiao)


# Gráfico de barras agrupado para comparar, por região, o percentual de escolas com IN_INTERNET versus o percentual com IN_INTERNET_ALUNOS.
plt.figure(figsize=(12, 7))
# Calcular as médias para ambas as colunas
media_internet = df_internet_aluno.groupby('NO_REGIAO')[['IN_INTERNET', 'IN_INTERNET_ALUNOS']].mean().reset_index()
# Transformar os dados para formato longo para facilitar o uso do seaborn
media_internet_melted = media_internet.melt(id_vars='NO_REGIAO', value_vars=['IN_INTERNET', 'IN_INTERNET_ALUNOS'], var_name='Tipo', value_name='Proporção')
# Plotar o gráfico de barras agrupado
sns.barplot(data=media_internet_melted, x='NO_REGIAO', y='Proporção', hue='Tipo')
plt.title('Proporção de Escolas com Internet e que Disponibilizam para Alunos, por Região', fontsize=16)
plt.xlabel('Região', fontsize=12)
plt.ylabel('Proporção', fontsize=12)

plt.ylim(0, 1) # Limitar o eixo y para melhor visualização
plt.legend(title='Tipo de Acesso', labels=['Com Internet', 'Internet para Alunos'])
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

Há uma clara disparidade regional, com Sul e Centro-Oeste apresentando melhores proporções, enquanto Nordeste e Norte ficam significativamente abaixo da média nacional.

Apesar de uma alta proporção de escolas brasileiras ter acesso à internet (representada pela barra 'Com Internet'), o gráfico revela uma lacuna persistente e preocupante: uma parcela considerável dessas escolas não disponibiliza a conexão para uso dos alunos. Esta discrepância é mais acentuada nas regiões Norte e Nordeste, onde o percentual de escolas com internet para alunos é de apenas 34% e 39%, respectivamente. Tal cenário indica que a mera existência de infraestrutura de internet não se traduz automaticamente em oportunidades de inclusão digital para os estudantes, apontando para desafios que vão além da conectividade básica, como a falta de equipamentos adequados, políticas de uso ou suporte técnico. As escolas estão conectadas, mas os alunos, em muitas delas, permanecem desconectados.
